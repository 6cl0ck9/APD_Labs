```python
import numpy as np
import pandas as pd
from timeit import default_timer as timer

FILENAME = "household_power_consumption.txt"
```

```python
# ---------------- Numpy ----------------
types = [("Date", "U10"), ("Time", "U8"), ("Global_active_power",
"float64"), ("Global_reactive_power", "float64"), ("Voltage",
"float64"), ("Global_intensity", "float64"), ("Sub_metering_1",
"float64"), ("Sub_metering_2", "float64"), ("Sub_metering_3",
"float64")]

start_np = timer()
np_data = np.genfromtxt(FILENAME, delimiter=';', dtype=types, 
                        missing_values='?', names=True, encoding='utf-8')
np_data = np_data[~np.isnan(np_data["Global_active_power"])]
end_np = timer()

print(f"Numpy load time: {end_np - start_np:.2f}s")
```

```python
# ---------------- Pandas ----------------
start_pd = timer()
df = pd.read_csv(FILENAME, sep=';', na_values='?', low_memory=False)
df.dropna(inplace=True)
end_pd = timer()
print(f"Pandas load time: {end_pd - start_pd:.2f}s")
```

1)Обрати всі записи, у яких загальна активна споживана потужність перевищує 5 кВт.

```python
def task1(df, np_data):
    start = timer()
    result_pd = df[df['Global_active_power'] > 5]
    end = timer()
    print(f"Pandas time: {end - start:.4f}s")
    display(result_pd.head())

    start = timer()
    result_np = np_data[np_data['Global_active_power'] > 5]
    end = timer()
    print(f"NumPy time: {end - start:.4f}s")
    print(result_np[:5])
task1(df, np_data)
```

2)Обрати всі записи, у яких вольтаж перевищую 235 В.

```python
def task2(df, np_data):
    start = timer()
    result_pd = df[df['Voltage'] > 235]
    end = timer()
    print(f"Pandas time: {end - start:.4f}s")
    display(result_pd.head())

    start = timer()
    result_np = np_data[np_data['Voltage'] > 235]
    end = timer()
    print(f"NumPy time: {end - start:.4f}s")
    print(result_np[:5])
task2(df, np_data)
```

3)Обрати всі записи, у яких сила струму лежить в межах 19-20 А, для них виявити ті, у яких пральна машина та холодильних споживають більше, ніж бойлер та кондиціонер.

```python
def task3(df, np_data):
    start = timer()
    result_pd = df[(df['Global_intensity'].between(19, 20)) & 
                   (df['Sub_metering_2'] > df['Sub_metering_3'])]
    end = timer()
    print(f"Pandas time: {end - start:.4f}s")
    display(result_pd.head())

    start = timer()
    mask = (np_data['Global_intensity'] >= 19) & \
           (np_data['Global_intensity'] <= 20) & \
           (np_data['Sub_metering_2'] > np_data['Sub_metering_3'])
    result_np = np_data[mask]
    end = timer()
    print(f"NumPy time: {end - start:.4f}s")
    print(result_np[:5])
task3(df, np_data)
```

4)Обрати випадковим чином 500000 записів (без повторів елементів вибірки), для них обчислити середні величини усіх 3-х груп споживання електричної енергії.

```python
def task4(df, np_data):
    start = timer()
    sample_pd = df.sample(n=500000, random_state=42)
    means_pd = sample_pd[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].mean()
    end = timer()
    print(f"Pandas time: {end - start:.4f}s")
    print(means_pd)

    start = timer()
    rng = np.random.default_rng(seed=42)
    indices = rng.choice(len(np_data), size=500000, replace=False)
    sample_np = np_data[indices]
    means_np = (
        sample_np['Sub_metering_1'].mean(),
        sample_np['Sub_metering_2'].mean(),
        sample_np['Sub_metering_3'].mean()
    )
    end = timer()
    print(f"NumPy time: {end - start:.4f}s")
    print(f"Means: Group 1: {means_np[0]}, Group 2: {means_np[1]}, Group 3: {means_np[2]}")
task4(df, np_data)
```

Обрати ті записи, які після 18-00 споживають понад 6 кВт за хвилину в середньому, серед відібраних визначити ті, у яких основне споживання електроенергії у вказаний проміжок часу припадає на пральну машину, сушарку, холодильник та освітлення (група 2 є найбільшою), а потім обрати кожен третій результат із першої половини та кожен четвертий результат із другої половини.

```python
def task5(df, np_data):
    start = timer()
    df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%d/%m/%Y %H:%M:%S')
    after_6 = df[(df['Datetime'].dt.hour >= 18) & (df['Global_active_power'] > 6)]
    mainly_2 = after_6[after_6['Sub_metering_2'] > after_6[['Sub_metering_1', 'Sub_metering_3']].max(axis=1)]
    half = len(mainly_2) // 2
    result_pd = pd.concat([mainly_2.iloc[:half:3], mainly_2.iloc[half::4]])
    end = timer()
    print(f"Pandas time: {end - start:.4f}s")
    display(result_pd.head())

    start = timer()
    df_np = pd.DataFrame(np_data)
    df_np['Datetime'] = pd.to_datetime(df_np['Date'] + ' ' + df_np['Time'], format='%d/%m/%Y %H:%M:%S')
    after_6_np = df_np[(df_np['Datetime'].dt.hour >= 18) & (df_np['Global_active_power'] > 6)]
    mainly_2_np = after_6_np[after_6_np['Sub_metering_2'] > after_6_np[['Sub_metering_1', 'Sub_metering_3']].max(axis=1)]
    half = len(mainly_2_np) // 2
    result_np = pd.concat([mainly_2_np.iloc[:half:3], mainly_2_np.iloc[half::4]])
    end = timer()
    print(f"NumPy time: {end - start:.4f}s")
    display(result_np.head())
task5(df, np_data)
```
